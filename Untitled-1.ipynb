```python
from pathlib import Path
import json
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
def load_jsonl(input_path: Path) ->  list[dict]:
    records: list[dict] = []
    with input_path.open("r", encoding="utf-8") as file:
        for line in file:
            record = json.loads(line)
            records.append(record)

    return records
```

```python
report_index_records = load_jsonl(
    Path("data/processed/report_index_records.jsonl")
)

note_index_records = load_jsonl(
    Path("data/processed/note_index_records.jsonl")
)
```

```python
all_index_records = (
    report_index_records
    + note_index_records
)
```

```python
print(all_index_records[20])
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1200,
    chunk_overlap = 200,
    separators= [
        "\n\n",
        "\n",
        ". ",
        " ",
        ""

    ]
)
```

```python
example_record = next(
    record
    for record in all_index_records
    if record["document_type"] == "internal note"
)
```

```python
chunks = text_splitter.split_text(
    example_record["content"]
)
```

```python
print(f"Número de chunks: {len(chunks)}")
```

```python
for i, chunk in enumerate(chunks):
    print(f"\n----- Chunk {i} -----")
    print(f"Longitud: {len(chunk)}")
    print(chunk[:200])
```

**empezamos generacion de nuevo dataset de chunks

```python
chunk_records = []

for record in all_index_records:
    document_type = record["document_type"]
    content = record['content']

    if document_type == "report":
        text_chunks = [content]
    else:
        text_chunks = text_splitter.split_text(content)

    for  chunk_index, chunk in enumerate(text_chunks):
        chunk_record = record.copy()
        chunk_record['content'] = chunk
        chunk_record['chunk_index'] = chunk_index

        source_stem = Path(record['source_file']).stem
        page_number = record.get('page_number', 1)

        chunk_record['chunk_id'] = (f'{source_stem}_p{page_number}_c{chunk_index}')
        
        chunk_records.append(chunk_record)
```

```python
#validacion

print(f"Registros antes: {len(all_index_records)}")
print(f"Chunks después: {len(chunk_records)}")
print(chunk_records[20])

# Unicidad

chunk_ids = [
    record["chunk_id"]
    for record in chunk_records
]

print(len(chunk_ids) == len(set(chunk_ids)))
```

```python
def save_jsonl(records: list[dict[str, str]],
               output_path: Path,
               ) -> None:
    output_path.parent.mkdir(parents=True, exist_ok=True)

    with output_path.open("w", encoding="utf-8") as file:
        for record in records:
            json_line = json.dumps(
                record,
                ensure_ascii=False,
            )
            file.write(json_line + "\n")
```

```python
save_jsonl(
    chunk_records,
    Path("data/processed/chunk_records.jsonl")
)
```
