```python
from pathlib import Path
import json
```

```python
def load_jsonl(input_path: Path) ->  list[dict]:
    records: list[dict] = []
    with input_path.open("r", encoding="utf-8") as file:
        for line in file:
            record = json.loads(line)
            records.append(record)

    return records
```

```python
report_index_records = load_jsonl(
    Path("data/processed/report_index_records.jsonl")
)

note_index_records = load_jsonl(
    Path("data/processed/note_index_records.jsonl")
)
```

```python
all_index_records = (
    report_index_records
    + note_index_records
)
```

```python
print(f"Anexos: {len(report_index_records)}")
print(f"Páginas de notas: {len(note_index_records)}")
print(f"Total indexable: {len(all_index_records)}")
```
