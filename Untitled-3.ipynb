```python
documents = [
    {"id": "A", "text": "Protocolo ante una intrusión"},
    {"id": "B", "text": "Procedimiento de evacuación"},
    {"id": "C", "text": "Acceso de personas no autorizadas"},
    {"id": "D", "text": "Avería de una cámara de seguridad"},
    {"id": "E", "text": "Horario del personal de limpieza"},
]
```

```python
embeddings = {
    "A": [0.9, 0.8, 0.1, 0.0, 0.1],
    "B": [0.1, 0.2, 0.9, 0.8, 0.0],
    "C": [0.8, 0.9, 0.1, 0.1, 0.0],
    "D": [0.3, 0.2, 0.1, 0.9, 0.8],
    "E": [0.0, 0.1, 0.2, 0.1, 0.9],
}
```

```python
vector_records = []

for document in documents:
    record = document.copy()
    record["embedding"] = embeddings[document["id"]]
    vector_records.append(record)
```

```python
import numpy as np

def cosine_similarity(vector_a, vector_b):
    a = np.array(vector_a, dtype=float)
    b = np.array(vector_b, dtype=float)
    numerator = np.dot(a, b)
    denominator = np.linalg.norm(a)* np.linalg.norm(b)

    return numerator/denominator
```
